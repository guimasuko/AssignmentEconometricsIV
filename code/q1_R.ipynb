```r
library(tidyverse)
library(readxl)
library(corrr)
library(ggcorrplot)
library(FactoMineR)
```

```r
directory = dirname(getwd())
directory
```

```r
input_path = paste(directory,'/data/returns.xlsx', sep = "")
df <- read_excel(input_path)
```

```r
factors_col = colnames(df)[2:17]
factors_col
```

```r
factors <- df %>%
select('dates', all_of(factors_col))
head(factors, 5)
```

```r
returns <- df %>%
select(-one_of(factors_col))
head(returns, 5)
```

```r
head(returns[,-1])
```

```r
standard_returns <- scale(returns[,-1])
head(standard_returns)
```

```r
data.pca <- princomp(standard_returns)
```
